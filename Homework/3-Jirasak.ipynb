```python
def find_maximum_subarray(A, start, end):
    if start == end :
        return start, end, A[start]
    else:
        mid = (start + end) // 2

        left_start, left_end, left_max = find_maximum_subarray(A, start, mid)
        right_start, right_end, right_max = find_maximum_subarray(A, mid + 1, end)
        cross_start, cross_end, cross_max = find_maximum_crossing_subarray(A, start, mid, end)

        if left_max >= right_max and left_max >= cross_max:
            return left_start, left_end, left_max
        elif right_max >= left_max and right_max >= cross_max:
            return right_start, right_end, right_max
        else:
            return cross_start, cross_end, cross_max


def find_maximum_crossing_subarray(A, start, mid, end):
    sum_left = -float('inf')
    sum_temp = 0
    for i in range(mid, start - 1, -1):
        sum_temp += A[i]
        if sum_temp >= sum_left:
            sum_left = sum_temp
            max_left = i
    
    sum_right = -float('inf')
    sum_temp = 0
    for j in range(mid + 1, end + 1):
        sum_temp += A[j]
        if sum_temp >= sum_right:
            sum_right = sum_temp
            max_right = j
    
    return max_left, max_right, sum_left + sum_right
```

```python
A = [3, 4, -2, 3, -10, 32, 4, -11, 7, -3, 2]
low = 0
high = len(A) - 1
start_index, end_index, max_sum = find_maximum_subarray(A, low, high)
print(f'Maximum subarray starts at index {start_index}, ends at index {end_index}, with maximum sum of {max_sum}')
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt


choices = [10, 100, 1_000, 10_000, 100_000, 1_000_000]
time_taken_history = []
for choice in choices:
    A = [0] * choice
    low = 0
    high = len(A) - 1

    start = time.time()
    find_maximum_subarray(A, low, high)
    end = time.time()
    time_taken = end - start
    print(choice, 'finished')
    time_taken_history.append(time_taken)
print(time_taken_history)
```

```python
plt.plot(time_taken_history)
plt.xticks(ticks=range(len(choices)), labels = choices);
plt.title('Number of element vs Actual time taken')
plt.xlabel('Number of element')
plt.ylabel('Actual time taken')
plt.show()
```

```python
x = np.array(choices)
y_theory = x * np.log(x)
print(y_theory)
```

```python
plt.plot(y_theory, c='orange')
plt.xticks(ticks=range(len(choices)), labels = choices);
plt.title('Number of element vs Time taken in theory')
plt.xlabel('Number of element')
plt.ylabel('Time taken in theory')
plt.show()
```

It can be seen that the both graph look similar to each other in term of shape, the different is that it's in the different scale, we will scale down the time taken in theory to be able to compare with the actual time taken

```python
# Find the constant to scale by comparing the max of each y 

scale_term = max(y_theory) / max(time_taken_history)
print(f'`y_theory` will be scaled down by {scale_term:.2f} times ')

c = 1 / scale_term

y_theory_scaled = [y * c for y in y_theory]
```

```python
plt.plot(time_taken_history, label = 'Actual time taken')
plt.plot(y_theory_scaled, label = 'Time taken in theory (scaled)')
plt.legend()
plt.xticks(ticks=range(len(choices)), labels = choices);
plt.title('Number of element vs Time taken')
plt.xlabel('Number of element')
plt.ylabel('Time taken')
plt.show()
```

It matched almost perfectly, we can now confirm that our algorithm take O(nlgn) time according to the theory
