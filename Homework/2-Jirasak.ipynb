```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.logspace(0,15,100,dtype=np.float32) 
y_1 = np.ones(len(X))
y_n = X
y_n2 = X**2
y_sqrt = np.sqrt(X)
y_nlogn = X * np.log(X)
y_n3 = X**3
y_exp = np.exp(X)
```

```python
fig = plt.figure(figsize=(12,8))
plt.xscale('log')
plt.yscale('log')
plt.plot(X, y_1 , label='$O(1)$')
plt.plot(X, y_n , label='$O(n)$')
plt.plot(X, y_n2, label='$O(n^2$)')
plt.plot(X, y_sqrt, label='$O(\sqrt{n}$)')
plt.plot(X, y_nlogn, label='$O(nlog(n))$')
plt.plot(X, y_n3, label='$O(n^3)$')
plt.plot(X, y_exp, label='$O(2^n)$')
plt.xlim(left=1,)
plt.ylim(bottom=1,)

plt.legend(prop={'size': 15}, loc='best')
plt.title('Visual comparison of each growth function')
```
